```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('amazon.csv')
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# drop the rows with null values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df['rating'].unique()
```

```python
df['rating'].replace('|','1.0',inplace=True)
```

```python
df['rating'].unique()
```

```python
df['rating'] = df['rating'].astype(float)
```

```python
df['rating'].unique()
```

```python
df['category'].unique()
```

```python
df.groupby(df['category'])['rating'].mean()
```

```python
df['rating_count'].unique()
```

```python
df['rating_count'] = df['rating_count'].str.replace(',','').astype(int)
```

```python
df['rating_count']
```

```python
df.info()
```

```python
df.groupby('category')['rating_count'].sum().sort_values(ascending=False)
```

```python
df['actual_price'].unique()
```

```python
df['actual_price'] = df['actual_price'].str.replace('₹','')
```

```python
df['actual_price'] = df['actual_price'].str.replace(',','').astype(float)
```

```python
df.info()
```

```python
df['discounted_price'].unique()
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹','')
```

```python
df['discounted_price'] = df['discounted_price'].str.replace(',', '').astype(float)
```

```python
df.info()
```

```python
df['actual_price']/df['discounted_price']
```

```python
df['actual_price'].sum()/df['discounted_price'].sum()
```

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(df['discounted_price'], color="blue", fill=True, label="Discounted Price", alpha=0.6)
sns.kdeplot(df['actual_price'], color="green", fill=True, label="Actual Price", alpha=0.6)

plt.title("Distribution of Discounted Prices vs Actual Prices", fontsize=16, weight='bold')
plt.xlabel("Price", fontsize=14)
plt.ylabel("Density", fontsize=14)

plt.legend(title="Price Type", fontsize=12, title_fontsize=13)

plt.tight_layout()
plt.show()
```

```python
df.columns
```

```python
df['discount_percentage'].unique()
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype(float)
```

```python
df.info()
```

```python
# # How does the average discount percentage vary across categories
# plt.figure(figsize=(20, 20))
# sns.barplot(x='category', y='discount_percentage', data=df, palette='viridis')
# plt.xticks(rotation=90)
# plt.xlabel('Category')
# plt.ylabel('Average Discount Percentage')
# plt.title('Average Discount Percentage by Category')

# Calculate average discount percentage by category
category_avg_discount = df.groupby('category')['discount_percentage'].mean().sort_values(ascending=False)

# Select the top 20 categories
top_categories = category_avg_discount.head(20).reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(
    x='discount_percentage',
    y='category',
    data=top_categories,
    palette='viridis',
    edgecolor=".2"
)

# Add titles and labels
plt.xlabel("Average Discount Percentage", fontsize=14, labelpad=10)
plt.ylabel("Category", fontsize=14, labelpad=10)
plt.title("Top 20 Categories by Average Discount Percentage", fontsize=16, weight='bold', pad=15)

# Fine-tune layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
a = df.groupby('category')['discount_percentage'].mean().sort_values(ascending=False)
a
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=a.index, y=a.values)
plt.xticks(rotation=90)
plt.xlabel('Category')
plt.ylabel('Average Discount Percentage')
plt.title('Average Discount Percentage by Category')
plt.show()
```
