```python
df['Annual Income'].value_counts()
```

```python
a,b,c,d=df['Annual Income'].quantile([0.25, 0.50,0.75,1])
a,b,c,d
```

```python
inc_bins= [0,a,b,c,d]
inc_labels = ['Low', 'Medium', 'High', 'Very High']
df['Income_Range'] = pd.cut(df['Annual Income'], bins=inc_bins,labels= inc_labels)
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Income_Range', y='Price', data=df, palette="coolwarm")
```

```python
inc_bins= [0,3,6,9,12]
inc_labels = ['q1', 'q2', 'q3', 'q4']
df['Quarter'] = pd.cut(df['Date'].str.split('/', expand=True)[0].astype(int), bins=inc_bins,labels= inc_labels)
```

```python
df.head()
```

```python
quaterly_sale = df.groupby('Quarter')['Price'].sum()
quaterly_sale
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=quaterly_sale.index,y=quaterly_sale, palette='Set2')
plt.ylabel('Total Sales')
plt.tight_layout()
```
