```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('spotify.csv')
df
```

```python
# What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram
sns.histplot(df['Popularity'], kde=True)
plt.title('Distribution of Popularity')
plt.xlabel('Popularity')
plt.show()
```

```python
df[['Popularity', 'Duration (ms)']].corr()
```

```python
sns.scatterplot(x='Popularity', y='Duration (ms)', data=df)
plt.title('Popularity vs Duration')
plt.show()
```

```python
df['Artist'].value_counts().sort_values(ascending=False)
```

```python
# Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot
plt.figure(figsize=(20, 6))
sns.countplot(x='Artist', data=df)
plt.title('Count of Tracks by Artist')
plt.xticks(rotation=90)
plt.show()
```

```python
#  What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each
df['Popularity'].sort_values(ascending=True).head(5)
```

```python
df['Popularity'].sort_values(ascending=True).head(5).values
```

```python
for pop in df['Popularity']:
  if pop in df['Popularity'].sort_values(ascending=True).head(5).values:
    print(df[df['Popularity'] == pop][['Artist', 'Track Name']])
```

```python
popular_artist = df.groupby('Artist')['Popularity'].sum().sort_values(ascending=False).head()
popular_artist
```

```python
for artist in popular_artist.index:
  print(artist, df[df['Artist'] == artist]['Popularity'].mean())
```

```python
df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False)
```

```python
df['Popularity'].sort_values(ascending=False).head(5)
```

```python
# For the top 5 most popular artists, what are their most popular tracks? List the track name for each artis
for artist in popular_artist.index:
  print(artist, df[df['Artist'] == artist].sort_values(by='Popularity', ascending=False).head(1)['Track Name'].values[0])
```

```python
# Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot
plt.figure(figsize=(20, 6))
sns.boxplot(x='Artist', y='Duration (ms)', data=df)
plt.title('Duration of Tracks by Artist')
plt.xticks(rotation=90)
plt.show()
```

```python
#  How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot
plt.figure(figsize=(20, 6))
sns.violinplot(x='Artist', y='Popularity', data=df)
plt.title('Distribution of Track Popularity by Artist')
plt.xticks(rotation=90)
plt.show()
```
